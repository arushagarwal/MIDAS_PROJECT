```python
import tweepy  #for utilising twitter api
import jsonlines  #for creating and realing JSON Lines file
from prettytable import PrettyTable #to present data in tabular format as mentioned in task
import csv  #Having an additional feature to present data in tabular format in CSV file

#Credentials for using twiiter api, consumer key and secret are already created whereas access
#one's are to be generated.
consumer_key = ""
consumer_secret = ""
access_key = ""
access_secret = ""


# A function to get authorisation and tweet data
def get_tweets(username):
    # Authorization to consumer key and consumer secret
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

    # Access to user's access key and access secret
    auth.set_access_token(access_key, access_secret)

    # Calling api
    api = tweepy.API(auth)

    # Twitter api only allows 200 tweets data per request
    #So in order to get later tweets a function is created which takes the oldest tweet id,
    #already found and then new query is made from that max to new 200 until the response
    #new tweets has no more data.
    alltweets=[]
    new_tweets = api.user_timeline(screen_name=username, count=200)
    alltweets.extend(new_tweets)
    oldest = alltweets[-1].id - 1  # The oldest tweet is towards the end of the list.
    # Negative index denotes iterating from the end side

    while len(new_tweets) > 0:
        # all subsiquent requests use the max_id param to prevent duplicates and get the next
        #set of values
        new_tweets = api.user_timeline(screen_name=username, count=200, max_id=oldest)

        # Extending the new tweets to the already existing one's
        alltweets.extend(new_tweets)

        # update the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1

    #Using JSONLINES to get a .jsonl file stored in the same folder
    with jsonlines.open('output.jsonl', mode='w') as writer:
     for tweet in alltweets:
      #  print(tweet._json)   #Printing the json as in twitter response json exists by name of _json object
        writer.write(tweet._json) #writing each found json line by line




def printInformation():  #Creating a function to get the relevant information by parsing the
                         #JSONL file ie. text,date,likes,retweets and images

    table=PrettyTable(['Index','TEXT','CREATED DATE','FAVOURITE COUNT','RETWEET COUNT','MEDIA COUNT'])
    i=0     #Above created instance of PrettyTable which lets us present the output data in table format
    result=[]  #The final result list
    with jsonlines.open('output.jsonl',mode='r') as reader:  #Reading JSONLines
        for data in reader:

            text=data['text'].encode('utf-8')
            date=data['created_at']
            favourite=data['favorite_count']
            retweet=data['retweet_count']
            media=' '
            try:
             media=len(data['extended_entities']['media'])  #The extended entities has an object media that has all url of native media

            except:
             media='None'

            tempdata=[i,text,date,favourite,retweet,media]  #Tempdata holds data for each tweet which is inserted
            #into the final result list index wise
            result.insert(i,tempdata)

            table.add_row([i,text,date,favourite,retweet,media])
            i=i+1


    header = ['Index', 'Text','Created Date','Favourite Count','Retweet Count','Media Count']
    with open('outputs.csv', 'w') as csvFile:  #Line 83 to 88 is the code for creating a
# CSV file of the data to represent in table format. We can comment these lines if only final output in only terminal is requred.
            writer = csv.writer(csvFile)
            writer.writerow(i for i in header)
            writer.writerows(result)


    print(table)  #This prints the data in a tabular format in console or terminal


# Driver code
if __name__ == '__main__':
    # Here goes the twitter handle for the user
    # whose tweets are to be extracted.
    get_tweets("midasIIITD")
    printInformation()  #This is used to read the stored JSONL. This is called seperately here
    # as there are two tasks one for writing and another for reading.
```
